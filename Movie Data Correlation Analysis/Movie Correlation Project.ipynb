```python
# Import the necessary libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab

plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
#Adjust the configuration of the plots created
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# ( Prepare the data) Read the file
df = pd.read_csv(r'C:\Users\Vandi\OneDrive\Desktop\movies.csv')
```

```python
# Overview of the data
df.head()
```

```python
# Process the data
# Let's clean the data
# Check the data for missing values

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# drop the corresponding rows with missing values
df = df.dropna()
```

```python
# Check the datatypes of our columns
df.dtypes
```

```python
# Change the datatype of budget column to int
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
# Since columns 'relased' and 'year' data is not matching create 'yearcorrect' column
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
# Order our data by 'gross' column in descending order
df=df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Visualize the data
# Scatter plot budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
df.head()
```

```python
# Budget vs gross plot using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# Let's look at the correlation

df_numeric = df.select_dtypes(include=[np.number])  # Select only numeric columns
df_numeric.corr(method='pearson')  # Calculate correlation on numeric columns
```

```python
correlation_matrix=df_numeric.corr(method='pearson') 

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```
